```python
import numpy as np
import json
import pandas as pd
import os
import sys
```

```python
currentDir = os.getcwd()
Filename='data/extract-paper.json'
currentFileCSV = currentDir +"\\" + Filename
with open(currentFileCSV,encoding='utf_8_sig', errors='ignore') as read_file:
    data = json.load(read_file)
```

```python
def getvalue(x):
    if(isinstance(x, dict)):
        return x['@value']
    if(isinstance(x, list)):
        return x[0]['@value']
        
```

```python
def partition(x):
    if(isinstance(x, list)):#check whether is list
        return [item.split('/')[-1] for item in x]
    if(isinstance(x, str)):
        return x.split('/')[-1]
```

```python
graph=pd.DataFrame(data['@graph'])
graph['@id']=graph['@id'].apply(lambda x:partition(x))
```

```python
graph.dropna(axis=1,how='all')
graph.describe()
```

```python
np.unique(graph['type'])
```

```python
basic=graph[['@id','type','label']]
basic.type=basic.type.apply(lambda x:partition(x))
basic.label=basic.label.apply(lambda x:getvalue(x))
basic.to_csv('csv/type.csv')
```

```python
graph.groupby(['type'])['@id'].count()
```

# Class

```python
Class=graph[graph['type']=='http://www.w3.org/2000/01/rdf-schema#Class']
Class=Class.dropna(axis=1,how='all')
Class['english_label']=Class['label'].apply(lambda x:x[0]['@value'])
Class['chinese_label']=Class['label'].apply(lambda x:x[1]['@value'])
del Class['type']
del Class['label']
Class=Class.set_index('@id')
Class.to_csv("csv/class.csv",encoding='utf_8_sig')
```
